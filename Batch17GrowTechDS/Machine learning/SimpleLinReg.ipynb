# Import All the essential library

```python
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Data Loading

```python
df = ps.read_csv('Salary Data.csv')
df
```

```python
new_df = df.drop(['Age',"Gender",'Education Level','Job Title'],axis = 1)
new_df
```

```python
# Find the null values in dataset
new_df.isnull().sum()
```

```python
# Fill null values with median value

new_df['Salary'].fillna(new_df['Salary'].median(),inplace=True)
new_df['Years of Experience'].fillna(new_df['Years of Experience'].median(),inplace=True)
```

```python
new_df.isnull().sum()
```

# split into dependent and independent

```python
X = np.array(new_df['Years of Experience']).reshape(-1,1) # independent data
y = np.array(new_df['Salary']) # Dependent data
```

```python
X
```

```python
y
```

# Split the dataset into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(X_train,y_train)
```

# Model Prediction

```python
y_test
```

# Model Prediction

```python
y_pred = LinReg.predict(X_test)
y_pred[:5]
```

```python
act_pred_df = ps.DataFrame({'Actual':y_test,
                            "Predicted":y_pred})
act_pred_df
```

# Model Evaluation

```python
LinReg.score(X_train,y_train)*100
```

```python
LinReg.score(X_test,y_test)*100
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_test,y_pred)
```
