# VECTOR DATABASE FAISS

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

Load the dataset

```python
loader = TextLoader("speech.txt")
document = loader.load()
document
```

Split the Text

```python
text_splitter = CharacterTextSplitter(chunk_size = 1000,
                                      chunk_overlap = 20)
docs = text_splitter.split_documents(document)
docs
```

Embedding and Vector Storing

```python
embedding = OllamaEmbeddings(model = 'mxbai-embed-large')
db = FAISS.from_documents(docs,embedding)
db
```

```python
query = "How does the speaker describe the desired output of the war?"
query_result = embedding.embed_query(query)
query_result
```

```python
db_result_output = db.similarity_search(query)
db_result_output[0].page_content
```

```python
doc_and_score = db.similarity_search_with_score(query)
doc_and_score
```

```python
# Saving and Loading
db.save_local("speech_db")
```

```python
new_db = FAISS.load_local('speech_db',
                          embedding,
                          allow_dangerous_deserialization=True)
query = "what is the main summery of this document?"
docss = new_db.similarity_search(query)
docss
```

# Chroma DB 

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
document = loader.load()
document
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000,
                                      chunk_overlap = 20)
docs = text_splitter.split_documents(document)
docs
```

```python
embedding = OllamaEmbeddings(model = 'mxbai-embed-large')
chroma_db = Chroma.from_documents(docs,embedding)
chroma_db
```

```python
query = "How does the speaker describe the desired output of the war?"
query_result = embedding.embed_query(query)
query_result
```

```python
db_result_output = chroma_db.similarity_search(query)
db_result_output[0].page_content
```

```python
# Save Chroma DB
Chroma.from_documents(documents=docs,embedding=embedding,
                      persist_directory = "Chroma_DB")
```

```python
# Load the Chroma_DB
chroma_db2 = Chroma(persist_directory="Chroma_DB",
       embedding_function=embedding)
```

```python
query = 'what does the speaker believe is the main reason the United States should enter the war?'
result = chroma_db2.similarity_search(query)
result[0].page_content
```

```python
result = chroma_db2.similarity_search_with_score(query)
result
```
