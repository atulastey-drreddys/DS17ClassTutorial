```python
# Import required library
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
```

```python
Response = requests.get(url)
Response    
```

```python
Response.content
```

```python
soup = BeautifulSoup(Response.text,'html.parser')
#soup
```

```python
# Extract the title
name = soup.find_all('a',class_ = 'title')
name_list = []
for i in name:
    title = i.text
    name_list.append(title.strip())
print(name_list)
```

```python
# Extract the price

price = soup.find_all('h4',class_ = 'price float-end card-title pull-right')
price_list = []
for i in price:
    prices = i.text
    price_list.append(prices.strip())
print(price_list)
```

```python
# Price Extraction
price = soup.find_all('h4',class_ = 'price float-end card-title pull-right')
price_list = []
for i in price:
    prices = i.text
    price_list.append(prices.strip())
print(price_list)
```

```python
# Extract the description
desc = soup.find_all('p',class_ = 'description card-text')
desc_list = []
for i in desc:
    description = i.text
    desc_list.append(description.strip())
print(desc_list)
```

```python
# Extract Review

review = soup.find_all('p',class_ = 'review-count float-end')
review_list = []
for i in review:
    review_ = i.text
    review_list.append(review_.strip())
print(review_list)
```

```python
# convert this information into dataframe

df = pd.DataFrame({
    "Title":name_list,
    "Price":price_list,
    "Description":desc_list,
    "Review":review_list
})
```
