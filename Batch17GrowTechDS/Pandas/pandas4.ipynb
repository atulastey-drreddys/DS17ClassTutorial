# Apply
1) 0-1000 --> Low
2) 1000-4000 --> Medium
3) Above 4000 --> High

```python
def Like(num):
    if num > 0 and num < 1000:
        return "Low"
    elif num >= 1000 and num < 4000:
        return "Medium"
    else:
        return "High"
```

```python
sme['Like_Category'] = sme['likes'].apply(Like)
```
