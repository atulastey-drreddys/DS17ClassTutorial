# Series

```python
l1 = ['Atul',"Rahul","Vinay","Raj"]
print(l1)
print(type(l1))
```

```python
ser = pd.Series(l1)
print(ser)
print(type(ser))
```

```python
l1 = ['Atul',"Rahul","Vinay","Raj"]
index = ["x",'y','z','a']
ser = pd.Series(l1,index=index)
print(ser)
print(type(ser))
```

```python
ser1 = pd.Series([54,643,345,65,])
print(ser1)
print(type(ser1))
```

```python
ll1 = [54,643,345,65,]
print(ll1)
```

```python
ser1 = pd.Series([54,643,345,65,],
                 index = ['Marker','Ink',"Duster","Pencil"])
print(ser1)
print(type(ser1))
print(ser1.dtype)
```

```python
ser1 = pd.Series([54,643,345,65,],
                 index = ['Marker','Ink',"Duster","Pencil"],
                 dtype=float)
print(ser1)
print(type(ser1))
print(ser1.dtype)
print(len(ser1))
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line
plt.figure(figsize=(9,3))
plt.plot(sme['likes'])
plt.show()
```

```python
# Bar chart 
plt.figure(figsize=(9,3))
plt.bar(x = sme['platform'],height=sme['likes'])
plt.show()
```

```python
# Bar chart 
plt.figure(figsize=(9,4))
color = ['green','orange','black']
plt.bar(x = sme['platform'],height=sme['likes'])
plt.xlabel("Platform")
plt.ylabel("Likes")
plt.title("Platform wise Like")
plt.show()
```

```python
plt.figure(figsize=(9,4))
color = ['green','orange','black']
sns.barplot(x = sme['platform'],y=sme['likes'],hue = sme['platform'])
plt.xlabel("Platform")
plt.ylabel("Likes")
plt.title("Platform wise Like")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,4))
color = ['green','orange','black']
sns.barplot(x = sme['post_type'],y=sme['likes'],)
plt.xlabel("Platform")
plt.ylabel("Likes")
plt.title("Platform wise Like")
plt.show()
```

```python
plt.figure(figsize=(9,4))
color = ['green','orange','black']
sns.barplot(x = sme['post_type'],y=sme['likes'],hue = sme['post_type'])
plt.xlabel("Platform")
plt.ylabel("Likes")
plt.title("Platform wise Like")
plt.show()
```

```python
plt.figure(figsize=(9,4))
#color = ['green','orange','black']
sns.barplot(x = sme['post_type'],y=sme['likes'],hue = sme['platform'])
plt.xlabel("Post Type")
plt.ylabel("Likes")
plt.title("Platform wise Like")
plt.show()
```
