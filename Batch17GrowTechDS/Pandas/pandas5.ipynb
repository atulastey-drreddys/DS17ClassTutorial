# Data Frame concat, join, merge

```python
df1 = pd.DataFrame({"A":[23,23,54],
                    "B":[435,67,54],
                    "C":[34,75,64]})
df2 = pd.DataFrame({"D":[273,283,544],
                    "e":[45,67,540],
                    "f":[324,750,90]})
```

```python
df1
```

```python
df2
```

```python

concat_row = pd.concat((df1,df2),axis = 0)
concat_row
```

```python
concat_row = pd.concat((df1,df2),axis = 0,ignore_index = True)
concat_row
```

```python
concat_col = pd.concat((df1,df2),axis = 1)
concat_col
```

```python
df3 = pd.DataFrame({"i":[2643,54],
                    "j":[45,54],
                    "k":[334,75]})
```

```python
three_df_concat = pd.concat((df1,df2,df3),axis = 1)
three_df_concat
```

```python
df1 = pd.DataFrame({"A":['x','u','y','z'],
                    "B":[435,67,54,53],
                    "C":[34,75,64,33]})
df2 = pd.DataFrame({"A":['x','y','z','v'],
                    "e":[45,67,540,90],
                    "f":[324,750,90,66]})
```

```python
df1
```

```python
df2
```

```python
merge = df1.merge(df2,on = ['A'])
merge
```

```python
merge_left = df1.merge(df2,on = ['A'],how = 'left')
merge_left
```

```python
merge_right = df1.merge(df2,on = ['A'],how = 'right')
merge_right
```

```python
merge_right = df1.merge(df2,on = ['A'],how = 'outer')
merge_right
```

```python
merge_inner = df1.merge(df2,on = ['A'],how = 'inner')
merge_inner
```

```python
merge_inner = df1.merge(df2,on = ['A'],how = 'cross',)
merge_inner
```

```python
result =pd.merge(df1,df2,how='cross')
result
```

```python
result =pd.merge(df1,df2,how='outer',on = 'A',validate='one_to_one')
result
```

```python
result =pd.merge(df1,df2,how='outer',on = 'A',validate='one_to_many')
result
```

```python
result =pd.merge(df1,df2,how='outer',on = 'A',validate='many_to_many')
result
```
